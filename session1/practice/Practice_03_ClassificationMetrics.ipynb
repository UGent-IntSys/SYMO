# Practical session: Evaluation metrics

```python
# Libraries

import pandas as pd
import matplotlib.pyplot as plt

from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score
from sklearn.metrics import f1_score
# others
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve

from sklearn.model_selection import train_test_split
```

## Metrics

For each model predictions, compute the following metrics:
- Accuracy
- Precision
- Recall
- F1-sccore
- Confusion Matrix


```python
## Data loading
# Set the data path 
DATA_PATH="../data/predictions/"
y_test = pd.read_csv(DATA_PATH + 'y_test.csv')

# Model DecisionTree: Predictions
y_DT = pd.read_csv(DATA_PATH + 'decisionTree_predictions.csv')
y_DT_prob = pd.read_csv(DATA_PATH + 'decisionTree_predictProb.csv')
```

```python
## Evalute performance 
accuracy_DT = accuracy_score(y_test,y_DT)
print("Accuracy score:", accuracy_DT)
```

```python
## Where the model does mistakes
cm = confusion_matrix(y_test, y_DT)
#tn, fp, fn, tp = cm.ravel()
disp = ConfusionMatrixDisplay(cm)
disp.plot()
```

## Receiver operating characteristic (ROC) curve

    Hint: Use the predict probablities (*predict_prob.csv)

```python
# Plot ROC curve
plt.figure(0).clf()

# Compute false positive rate, true positive rate and AUC (Area under the curve)
fpr, tpr, _ = roc_curve(y_test, y_DT_prob)
auc = round(roc_auc_score(y_test, y_DT_prob), 4)
plt.plot(fpr,tpr,label="Decision Tree, AUC="+str(auc))

#add legend
plt.legend()
```
